```python
# Part A

def times_three(x):
    return x*3

# We recognize that the defined function takes a parameter and returns that value multiplied by 3

n=2

# n is now defined as the value 2 and stored in memory

times_three(n)

#we call the function "times_three" and consume the stored variable n which is 2
#the function executes and returns 2*3 = 6
#however...

n

#the last line which Python will output is not times_three(n) but rather solely n
#this means that the final answer should be the stored value from line 8
```

```python
# Part B

#This type of Python tracing question can be easier to solve using pen/paper
#Using a pen/paper method will let you keep track of any changes to the stored memory values line by line
#Let's try to go through the question step-by-step 

x = 2
y = 3

# x is now stored in memory as the value 2
# y is now stored in memory as the value 3

t = x 
# since x =2 => t = 2
x = y
# The variable y is stored in memory as the value 3
# The variable x will now be replaced with the y value which is 3
y = t

# The variable t has a stored value of 2
# Variable y will now be replaced with the t value which is 2
y

# The final call is to variable y which is now 2
```
